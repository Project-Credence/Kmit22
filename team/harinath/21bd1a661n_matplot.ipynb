```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30])
y=np.array([40,25,60])
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,3,4,5,7])
y=np.array([40,25,60,5,67,8,3])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatterplot")
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40])
plt.pie(x,labels=["10","20","30","40"])
plt.legend(title="pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,3 * np.pi,0.1)
y=np.sin(x)
plt.title("sine wave")
plt.plot(x,y)
plt.show()
```
